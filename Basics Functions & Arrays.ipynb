# 1. Array Operations

**(A).Swapping Program :-**

```python
# This program swap two elements in an array based on the indices provided by the user.
def swap_element(arr, index_1, index_2):
  print("Original Array:",arr)
  arr[index_1], arr[index_2] = arr[index_2], arr[index_1]
  print(f"Array after swapping index {index_1} and {index_2}:")
  return arr

array = [12,13,14,15,16,17,18,19]
swap_element(array,1,3)
```

```python
# Time Complexity : O(1)
# Analysis :- Accessing array elements by their index and swapping them takes constant amount of time, regardless of how large the array is.
```

**(B). Splitting an Array :-**

```python
# This program split an array into two separate arrays at specific given position.
def split_arr(arr, split_index):
  print("Original Array:",arr)
  part_1 = arr[:split_index]
  part_2 = arr[split_index:]
  print("Part_1:",part_1)
  print("Part_2:",part_2)
  return "After splitting array:",part_1, part_2

array = [1,2,3,4,5,6,7,8]
split_arr(array,4)
```

```python
# Time Complexity : O(N)
# Analysis :- Slicing the array requires iterating through the elements to copy them into the new subarrays. Therefore, the time taken scales linearly with N, where N is the number of elements being copied.
```

# 2. Calculator Program (Using user-define functions)

```python
# This program uses separate user-defined functions for each mathematicals operations.
def add(a,b):
  return a + b

def subtract(a,b):
  return a - b

def multiply(a,b):
  return a * b

def divide(a,b):
  if b == 0:
    return "Erro! Division by zero"
  return a / b

def Calculator(num_1, operation, num_2):
  if operation == "+":
    return add(num_1, num_2)
  elif operation == "-":
    return subtract(num_1, num_2)
  elif operation == "*":
    return multiply(num_1, num_2)
  elif operation == "/":
    return divide(num_1, num_2)
  else:
    return "Invalid operation !"

cal = Calculator(5,"+",2)
print("Addition: 5 + 2 =", cal)
```

```python
# Time Complexity : O(1)
# Analysis :- Arithmetic operations like addition, subtraction, multiplication and division are primitive operations that execute in constant time.
```

# (3). Simple Programs

**(A). Calculate Discount :-**

```python
# This program calculates final price after applying a percentage discount.
def calculate_discount(original_price, discount_percentage):
  discount_amount = (original_price * discount_percentage) / 100
  final_price = original_price - discount_amount

  print("Original Price:", original_price)
  print(f"Discount: {discount_percentage}% = RS.{discount_amount}")
  print("Final Amount:")
  return final_price

calculate_discount(7000, 12)
```

```python
# Time Complexity : O(1)
# Analysis :- This calculation involves a fixed number of basic math operations (multiplication, division and subtraction), which take constant time.
```

**(B). Calculate Grades :-**

```python
# This program assigns a letter grade based on a student's marks.
def calculate_grade(marks):
  if marks >= 90:
    grade = "A"
  elif marks >= 80:
    grade = "B"
  elif marks >= 70:
    grade = "C"
  elif marks >= 60:
    grade = "D"
  else:
    grade = "F"

  print(f"Marks: {marks} :- Grade: {grade}")


calculate_grade(79)
```

```python
# Time Complexity : O(1)
# Analysis :- This program uses conditional if-elif-else statements. The conditions are evaluated sequentially, and it takes a maximum of 5 comparisons to find the grade, which evaluates to constant time.
```

**(C). Odd/Even Check :-**

```python
# This program checks whether a given integer is odd and even.
def check_even_odd(number):
  if number % 2 == 0:
    print(f"{number} : is an Even number.")
  else:
    print(f"{number} : is an Odd number.")

check_even_odd(55)
```

```python
# Time Complexity : O(1)
# Analysis :- The modulas operator (%) and the equality check (==) are basic operations evaluated in constant time.
```
